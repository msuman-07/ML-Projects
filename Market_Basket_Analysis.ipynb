```python
dataset = pd.read_csv('grocery_dataset.csv')
print(dataset.shape)
print(dataset.head(5))
```

**Data Pre-Processing**

```python
transactions = []
for i in range(0, len(dataset)):
    transactions.append([str(dataset.values[i, j]) for j in range(0, 5)])
```

**Training APRIORI**

```python
!pip install apyori
```

```python
from apyori import apriori
model = apriori (transactions = transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

**RESULT**

```python
results = list(model)
results
```

**Results in Dataframe**

```python
lhs = [tuple(result[2][0][0])[0] for result in results]
rhs = [tuple(result[2][0][1])[0] for result in results]
supports = [result[1] for result in results]
confidences = [result[2][0][2] for result in results]
Lifts = [result[2][0][3] for result in results]
resultsinDataFrame = pd.DataFrame(zip(lhs, rhs, supports, confidences, Lifts), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
resultsinDataFrame
```
