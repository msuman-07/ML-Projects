# Import Libraries

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

# Load Dataset

```python
dataset = pd.read_csv("House Price Dataset.csv")
```

# Load Summarize

```python
print(dataset.shape)
print(dataset.head(5))
```

# Visualize Dataset

```python
plt.xlabel('Area')
plt.ylabel('Price')
plt.scatter(dataset.area, dataset.price, color='red', marker='*')
```

# Segregate Datset into I/P-X & O/P-Y

```python
X = dataset.drop('price', axis='columns')
X
```

```python
Y= dataset.price
Y
```

# Training Dataset using Linear Regression

```python
model = LinearRegression()
model.fit(X,Y)
```

# Predicted Price for Land sq. Feet of custom values

```python
x=40000
LandAreainSqFt = pd.DataFrame([x])
PredictedmodelResult = model.predict(LandAreainSqFt)
print(PredictedmodelResult)
```

# Theory Calculation
Y =m*X+b (m is coefficient and b is intercept)

Coefficient-m

```python
m=model.coef_
print(m)
```

Intercept-b

```python
b=model.intercept_
print(b)
```

# Y = mx+b
x is Independent varibal

```python
y = m*x + b
print("The Price of {0} Square feet Land is: {1}".format(x,y[0]))
```
